```python
url = "https://www.tripadvisor.com.sg/Hotel_Review-g294265-d301583-Reviews-Raffles_Hotel_Singapore-Singapore.html"

```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

driver = webdriver.Chrome()
driver.get(url)
import time 

f = open("urls.txt", "a")

for page_no in range(10):
    print (page_no)

    time.sleep(2)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight/3);")
    time.sleep(5)

    # assert "Python" in driver.title
    elems = driver.find_elements_by_tag_name("a")
    for e in elems:
        if (e != None):
            _url = e.get_attribute("href")
            if _url != None:
                if "ShowUserReviews" in _url or "showuserreviews" in _url:
                    print _url
                    f.write(_url + "\n")
            

    time.sleep(5)

    elem = driver.find_element_by_css_selector('div > .pagination > span.next')

    elem.click()
    time.sleep(5)


driver.close()
f.close()
```
